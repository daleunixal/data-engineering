```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Процессинг первого датасета


```python
df = pd.read_csv('./data/opt/[0]selected_dataset.csv')

df['date'] = pd.to_datetime(df['date'], format='%Y%m%d')

# График 1: Линейный график для числового столбца
plt.figure(figsize=(10, 6))
sns.lineplot(x='date', y='number_of_game', data=df)
plt.title('Line Plot of Number of Game over Time')
plt.xlabel('Date')
plt.ylabel('Number of Game')
plt.show()

# График 2: Гистограмма для числового столбца
plt.figure(figsize=(10, 6))
sns.histplot(df['number_of_game'], bins=10, kde=True)
plt.title('Histogram of Number of Game')
plt.xlabel('Number of Game')
plt.ylabel('Frequency')
plt.show()

# График 3: Круговая диаграмма для категориального столбца
plt.figure(figsize=(8, 8))
df['day_of_week'].value_counts().plot.pie(autopct='%1.1f%%')
plt.title('Pie Chart of Day of Week')
plt.ylabel('')
plt.show()

# График 4: График разброса для двух числовых столбцов
plt.figure(figsize=(10, 6))
sns.scatterplot(x='v_game_number', y='v_score', data=df)
plt.title('Scatter Plot of Visitor Game Number vs. Visitor Score')
plt.xlabel('Visitor Game Number')
plt.ylabel('Visitor Score')
plt.show()

# График 5: Ящик с усами для числового столбца
plt.figure(figsize=(10, 6))
sns.boxplot(x='h_league', y='h_game_number', data=df)
plt.title('Boxplot of Home Game Number by Home League')
plt.xlabel('Home League')
plt.ylabel('Home Game Number')
plt.show()
```

Процессинг второго датасета

```python
df_cars = pd.read_csv('./data/opt/[1]selected_dataset.csv')

# Преобразование типов данных (если необходимо)
df_cars['firstSeen'] = pd.to_datetime(df_cars['firstSeen'])
df_cars['lastSeen'] = pd.to_datetime(df_cars['lastSeen'])

# График 1: Линейный график для числового столбца (например, msrp)
plt.figure(figsize=(10, 6))
sns.lineplot(x='firstSeen', y='msrp', data=df_cars)
plt.title('Line Plot of MSRP over Time')
plt.xlabel('First Seen Date')
plt.ylabel('MSRP')
plt.show()

# График 2: Гистограмма для числового столбца (например, mileage)
plt.figure(figsize=(10, 6))
sns.histplot(df_cars['mileage'], bins=20, kde=True)
plt.title('Histogram of Mileage')
plt.xlabel('Mileage')
plt.ylabel('Frequency')
plt.show()

# График 3: Круговая диаграмма для категориального столбца (например, isNew)
plt.figure(figsize=(8, 8))
df_cars['isNew'].value_counts().plot.pie(autopct='%1.1f%%')
plt.title('Pie Chart of isNew')
plt.ylabel('')
plt.show()

# График 4: Ящик с усами для числового столбца (например, askPrice)
plt.figure(figsize=(10, 6))
sns.boxplot(x='isNew', y='askPrice', data=df_cars)
plt.title('Boxplot of Ask Price by isNew')
plt.xlabel('Is New')
plt.ylabel('Ask Price')
plt.show()

# График 5: Countplot для категориального столбца (например, color)
plt.figure(figsize=(12, 6))
sns.countplot(x='color', data=df_cars, order=df_cars['color'].value_counts().index)
plt.title('Countplot of Car Colors')
plt.xlabel('Color')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

Процессинг третьего датасета

```python
df_flights = pd.read_csv('./data/opt/[2]selected_dataset.csv')

# Преобразование типов данных (если необходимо)
df_flights['SCHEDULED_DEPARTURE'] = pd.to_datetime(df_flights['SCHEDULED_DEPARTURE'])

# График 1: Линейный график для временного ряда
plt.figure(figsize=(10, 6))
sns.lineplot(x='SCHEDULED_DEPARTURE', y='FLIGHT_NUMBER', data=df_flights)
plt.title('Line Plot of Flight Number over Scheduled Departure Time')
plt.xlabel('Scheduled Departure Time')
plt.ylabel('Flight Number')
plt.show()

# График 2: Гистограмма для дня недели вылета
plt.figure(figsize=(8, 6))
sns.histplot(df_flights['DAY_OF_WEEK'], bins=7, kde=False)
plt.title('Histogram of Day of Week for Departures')
plt.xlabel('Day of Week')
plt.ylabel('Frequency')
plt.show()

# График 3: Круговая диаграмма для авиакомпаний
plt.figure(figsize=(10, 10))
df_flights['AIRLINE'].value_counts().plot.pie(autopct='%1.1f%%')
plt.title('Pie Chart of Airlines')
plt.ylabel('')
plt.show()

# График 4: Ящик с усами для номера рейса
plt.figure(figsize=(10, 6))
sns.boxplot(x='AIRLINE', y='FLIGHT_NUMBER', data=df_flights)
plt.title('Boxplot of Flight Number by Airline')
plt.xlabel('Airline')
plt.ylabel('Flight Number')
plt.xticks(rotation=45)
plt.show()

# График 5: Countplot для месяца вылета
plt.figure(figsize=(8, 6))
sns.countplot(x='MONTH', data=df_flights)
plt.title('Countplot of Departures by Month')
plt.xlabel('Month')
plt.ylabel('Count')
plt.show()
```

Процессинг четвертого датасета

```python
df_jobs = pd.read_csv('./data/opt/[3]selected_dataset.csv')

# График 1: Круговая диаграмма для распределения заявленных навыков
key_skills_counts = df_jobs['key_skills'].str.split(',').explode().str.strip().value_counts()
plt.figure(figsize=(12, 12))
key_skills_counts.plot.pie(autopct='%1.1f%%')
plt.title('Pie Chart of Key Skills Distribution')
plt.ylabel('')
plt.show()

# График 2: Countplot для опыта работы
plt.figure(figsize=(10, 6))
sns.countplot(x='experience_name', data=df_jobs)
plt.title('Countplot of Experience Levels')
plt.xlabel('Experience Level')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()

# График 3: Гистограмма для длины описания вакансии
df_jobs['description_length'] = df_jobs['description'].apply(len)
plt.figure(figsize=(12, 6))
sns.histplot(df_jobs['description_length'], bins=30, kde=True)
plt.title('Histogram of Job Description Length')
plt.xlabel('Description Length')
plt.ylabel('Frequency')
plt.show()

# График 4: Ящик с усами для длины описания вакансии по дням недели
df_jobs['day_of_week'] = pd.to_datetime(df_jobs['created_at']).dt.day_name()
plt.figure(figsize=(12, 6))
sns.boxplot(x='day_of_week', y='description_length', data=df_jobs)
plt.title('Boxplot of Job Description Length by Day of Week')
plt.xlabel('Day of Week')
plt.ylabel('Description Length')
plt.show()

# График 5: Countplot для специализаций
specializations_counts = df_jobs['specializations'].str.split().explode().str.strip().value_counts()
plt.figure(figsize=(12, 6))
specializations_counts.plot.bar()
plt.title('Countplot of Job Specializations')
plt.xlabel('Specialization')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

Процессинг пятого датасета

```python
df_space_objects = pd.read_csv('./data/opt/[4]selected_dataset.csv')

# График 1: Гистограмма для диаметра космических объектов
plt.figure(figsize=(10, 6))
sns.histplot(df_space_objects['diameter'], bins=20, kde=True)
plt.title('Histogram of Space Objects Diameter')
plt.xlabel('Diameter')
plt.ylabel('Frequency')
plt.show()

# График 2: Рассеяние для характеристик космических объектов (H vs. diameter)
plt.figure(figsize=(10, 6))
sns.scatterplot(x='H', y='diameter', data=df_space_objects)
plt.title('Scatter Plot of Space Objects Characteristics')
plt.xlabel('H (Absolute Magnitude)')
plt.ylabel('Diameter')
plt.show()

# График 3: Ящик с усами для типа космических объектов
plt.figure(figsize=(12, 6))
sns.boxplot(x='name', y='diameter', data=df_space_objects)
plt.title('Boxplot of Space Objects Diameter by Type')
plt.xlabel('Space Object Type')
plt.ylabel('Diameter')
plt.xticks(rotation=45)
plt.show()

# График 4: Линейный график для характеристики H (Абсолютной звездной величины) по времени
plt.figure(figsize=(12, 6))
sns.lineplot(x='spkid', y='H', data=df_space_objects)
plt.title('Line Plot of Absolute Magnitude (H) over Time')
plt.xlabel('Space Object ID')
plt.ylabel('Absolute Magnitude (H)')
plt.show()

# График 5: Круговая диаграмма для категории NEO (Near-Earth Object)
neo_counts = df_space_objects['neo'].value_counts()
plt.figure(figsize=(8, 8))
neo_counts.plot.pie(autopct='%1.1f%%', labels=neo_counts.index, colors=['skyblue', 'lightcoral'])
plt.title('Pie Chart of Near-Earth Objects (NEO)')
plt.ylabel('')
plt.show()
```

Процессинг шестого датасета

```python
df_reviews = pd.read_csv('./data/opt/[5]selected_dataset.csv')

# График 1: Гистограмма для рейтинга обзоров
plt.figure(figsize=(10, 6))
sns.histplot(df_reviews['review_rating'], bins=5, kde=False)
plt.title('Histogram of Review Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()

# График 2: Ящик с усами для числа лайков обзоров
plt.figure(figsize=(10, 6))
sns.boxplot(x='review_likes', data=df_reviews)
plt.title('Boxplot of Review Likes')
plt.xlabel('Number of Likes')
plt.show()

# График 3: Круговая диаграмма для распределения версий приложения
app_version_counts = df_reviews['author_app_version'].value_counts().head(5)
plt.figure(figsize=(10, 10))
app_version_counts.plot.pie(autopct='%1.1f%%')
plt.title('Top 5 App Versions Distribution')
plt.ylabel('')
plt.show()

# График 4: Countplot для дней недели публикации обзоров
df_reviews['review_timestamp'] = pd.to_datetime(df_reviews['review_timestamp'])
df_reviews['day_of_week'] = df_reviews['review_timestamp'].dt.day_name()
plt.figure(figsize=(10, 6))
sns.countplot(x='day_of_week', data=df_reviews, order=['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])
plt.title('Countplot of Reviews by Day of Week')
plt.xlabel('Day of Week')
plt.ylabel('Count')
plt.show()

# График 5: Countplot для месяцев публикации обзоров
df_reviews['month'] = df_reviews['review_timestamp'].dt.month_name()
plt.figure(figsize=(10, 6))
sns.countplot(x='month', data=df_reviews, order=['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December'])
plt.title('Countplot of Reviews by Month')
plt.xlabel('Month')
plt.ylabel('Count')
plt.show()
```
